### 1.Read data

```python
import pandas as pd 
import numpy as np

SA1_file = "E:/GovHack/GovHack2021/data/2033055001 - sa1 indexes.xls"
df_sa1 = pd.read_excel(SA1_file)

xls = pd.ExcelFile(SA1_file)
df = xls.parse('Table 1', skiprows=5, index_col=None, na_values=['NA'])
df = df.replace('-', np.nan)
columns1 = ["SA1_7code", "SA1_11code", 
           "Score_disadvantage", "Decile_disadvantage", 
           "Score_advantage", "Decile_advantage", 
           "Score_economic", "Decile_economic", 
           "Score_occupation", "Decile_occupation",
           "population"]
df.columns = columns1
df.drop(columns=["SA1_11code"], inplace=True)
df.drop(df.tail(2).index, inplace=True)
df = df.set_index("SA1_7code")
df = df.fillna(df.mean())
df.head()
```

```python
print(len(df[df['Decile_occupation']==5.500009069307649]))
df.loc[df['Decile_occupation']==5.500009069307649, 'Decile_occupation'] = 5.5
print(len(df[df['Decile_occupation']==5.500009069307649]))
```

```python
df_norm = pd.get_dummies(df, columns=["Decile_disadvantage","Decile_advantage", "Decile_economic", "Decile_occupation"])
df_norm.head()
```

### 2.Kmeans 

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
from sklearn import datasets
```

```python
km = KMeans(n_clusters=10)
km.fit(df_norm)
km.predict(df_norm)
labels = km.labels_
```

```python
df['KMeans'] = labels.tolist()
df.head()
```

### 3.K-prototypes

```python
from kmodes.kprototypes import KPrototypes

kproto = KPrototypes(n_clusters=10, init='Cao')
clusters = kproto.fit_predict(df, categorical=[1, 3, 5, 7])

df['kprototypes'] = clusters.tolist()
df.head()
```
